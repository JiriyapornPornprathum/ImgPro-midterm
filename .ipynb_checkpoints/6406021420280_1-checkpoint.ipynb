```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(6.4*5, 4.8*5), constrained_layout=False)

img_c1 = cv2.imread("period_input.jpg", 0)
H = np.ones(img_c1.shape[:2])
H[150:190,230:250]=0
H[290:310,230:250]=0
H[210:240,150:180]=0
H[100:130,310:340]=0
H[210:230,310:340]=0
H[350:370,310:340]=0
H[290:310,400:430]=0
H[150:180,400:430]=0
H[210:240,470:500]=0
img_c2 = np.fft.fft2(img_c1)
img_c3 = np.fft.fftshift(img_c2)
img_cH = H*img_c3
img_c4 = np.fft.ifftshift(img_cH)
img_c5 = np.fft.ifft2(img_c4)


plt.figure(figsize = (15, 15)) , plt.subplot(221), plt.imshow(img_c1, "gray"), plt.title("1. Original Image")
plt.subplot(222), plt.imshow(np.log(1+np.abs(img_c2)), "gray"), plt.title("2. Spectrum")
# plt.subplot(143), plt.imshow(np.log(1+np.abs(img_c3)), "gray"), plt.title("Centered Spectrum")
plt.subplot(223), plt.imshow(np.log(1+np.abs(img_cH)), "gray"), plt.title("3. Filtered Spectrum")
plt.subplot(224), plt.imshow(np.abs(img_c5), "gray"), plt.title("4. Processed Image")

plt.show()
```
