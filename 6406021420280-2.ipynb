```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
img = cv2.imread('lowcon.png',0)
```

```python
def pixelVal(pix, r1, s1, r2, s2):
    if (0 <= pix and pix <= r1):
        return (s1 / r1)*pix
    elif (r1 < pix and pix <= r2):
        return ((s2 - s1)/(r2 - r1)) * (pix - r1) + s1
    else:
        return ((255 - s2)/(255 - r2)) * (pix - r2) + s2
```

```python
r1 = 140
s1 = 0
r2 = 200
s2 = 250
pixelVal_vec = np.vectorize(pixelVal)
contrast_stretched = pixelVal_vec(img, r1, s1, r2, s2)
```

```python
equ = cv2.equalizeHist(img)
```

```python
titles = ['1. Original','2. Contrast_stretched','3. EqualizeHist']
images = [img, contrast_stretched, equ,]
for i in range(3):
    plt.figure(figsize = (17, 17))
    plt.subplot(1,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
    
# plt.figure(figsize = (16, 16)) , plt.subplot(131), plt.imshow(img, "gray"), plt.title("1. Original Image")
# plt.subplot(132), plt.imshow(contrast_stretched, "gray"), plt.title("2. Contrast_stretched")
# plt.subplot(133), plt.imshow(equ, "gray"), plt.title("3. EqualizeHist")
    
```
